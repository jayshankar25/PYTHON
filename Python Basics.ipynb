#### Bitwise Operator:
    These are used to perform bit operation. All the decimal value will be converted into binaary value And Bitwise operation will works on these bits such as shifting them left to right or converting value from 0 t0 1 etc.
    
       - & ( Bitwise AND)
       - | ( Bitwise OR)
       - ^ ( Bitwise Exclusive OR)
       - ~ (Bitwise Complement)
       - << (shift Left)
       - '>>' (shift Right)
       
       


##### Bitwise AND (&) :-
 It compares each bit of the first operand to the corressponding bit of the second operand. if both bits are 1 , the corresponding bit is set to 1.
 
 Example:- 
     
    a = 10   ------>  00001010
    b = 20   ------>  00010100
  ------------------------------
    a & b             00000000
    

```python
a = 10 
b = 20
print(a&b)
```

#### Bitwise OR:-
it takes two bit patterns of equal length. this result in each position is 0 if both are 0 while otherwise the result is 1

 Example:- 
     
    a = 10   ------>  00001010
    b = 20   ------>  00010100
  ------------------------------
    a | b             00011110

```python
a = 10 
b = 20
print(a|b)
```

#### Bitwise Exclusive OR
    It compare each bit of its first operand to the corresponding bit of second operand, If one bit is 0 and the other bit is 1 , the corresponding result is set to 1 . otherwise the corresponding result bits is set to 0 ( identical is 0)
    
Examples:-  
         
         a = 1010  
         b = 0011
         --------
       a^b = 1001

```python
a = 10
b = 20
print(a^b)
```

### For Loop with else:-
    A for loop have an optional else block as well. The else part is executed if items in the sequence used in for loop exhausts. Break statement can be used to stop the for loop. in such case  the else part is ignored.
    
    - Syntax:-
         
         for <variable> in <sequence>:
             < statement >
             < statement >
         else:
             < statement >
             < statement >

Example:-

```python
for i in range(5):
    print(i)
else:
    print("else part")
    print(i)
```

```python
for i in range(5):
    print(i)
    if i == 2:
        break
else:
    print("else part")
    print(i)
```

#### continue:-
    - the continue keyword statement is used to tell python interpreter to skip the rest of the statement in the current loop block and to continue to the next iteration if the loop
    
                            or 
                            
    - a continue statement jumps to the next itration of the loop skipping any code in between
    
    syntax:
    
        1. continue
        
        2. for variable_name in sequence:
                statement-1
                statement-2
                
                if expression :
                    continue
                    statement-3
                    statement-4

Examples:-

```python
for i in range(10):
    print(i)
    if i == 5:
        continue
```

```python
for i in range(10):  
    if i == 5:
        print(i)
        continue
```

```python
for i in range(10):  # 0, 1, 2, 3, 4, 5, 6
    if i == 5:
        continue
        print(i)
```

```python
for i in range(10):
    if i == 5:                    
        continue                  
    else:
        print(i)
```

## While Loop:
    While loop is used to execute number of statements or body till the condition passed in while is True, once the condition is false, the control will come out of the loop
    
    syntax:-
    
        1. while <expression>:
            body
            
            or 
            
            statement

```python
x = 1

while x>0:
    print("positive")
    x = x + 1  # x += 1
```

```python
x = 1

while x>0:
    print(x)
    
    
    if x == 10:
        break  
    
    x = x + 1  # x += 1
```

```python
x = 1

while x <= 10:
    print(x)
    x = x + 1  # x += 1
    
print(x)
```

#### While with else:
    It is same as for with else.
    
    syntax:
          
          while <condition>:
              body
          else:
              body

```python
x = 1

while x <= 10:
    
    print(x)
    x = x + 1  # x += 1
else:  
    print(x)
```

### Python Nested Loop:
    python programming language allows to use one loop inside another loop
    
    syntax:
    
        for < first itertive varible > in outer loop:     # row
            [do something ]
            
            for < second itertive varible > in inner loop:  # column
                [do something ]
                
                
#### In nested loop , Find the following facts:-
    - x-axix always contain y-axis
    - y-axix always contain actual data
    - a row contains n number of columns
    - a column contains n number character (data)
    - outer loop represents x-axis (row)
    - inner loop represents y-axis (columns)
    - Mathematically we are representing width and height (x, y)

Examples:-

```python
for i in range(2):
    print(i)
    for j in range(4):
        print(j)
```

```python
for i in range(2):
    for j in range(4):
        print(i, j)
```

```python
for i in range(2):
    for j in range(4):
        if i == 0 and j == 3:
            print(i, j)
```

```python
for i in range(2):
    for j in range(4):
        if i == 0:
            print(i, j)
            
```

```python
for i in range(2):
    for j in range(4):
        if i == 1:
            print(i, j)
            
```

```python
for i in range(2):  #---> rows
    for j in range(4): # ----> columns
        if i == 1 and j > 0:
            print(i, j)
            
```

```python
number = 15

binary_number = int("{0:08b}".format(number))

flipped_binary_number = ~ binary_number

flipped_binary_number = flipped_binary_number + 1

str_twos_complement = str(flipped_binary_number)

twos_complement = int(str_twos_complement, 2)

print(twos_complement)
```
